```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
entro = []
for p in np.linspace(0.01,0.99):
    d = np.array([p,1-p])
    entro.append(-sum(d*np.log(d)))
```

```python
plt.plot(np.linspace(0.01,0.99), entro)
```

```python
entro = []
for p in np.linspace(0.01,0.99):
    d = np.array([p,(1-p)/2,(1-p)/2])
    entro.append(-sum(d*np.log(d)))
```

```python
plt.plot(np.linspace(0.01,0.99), entro)
```
