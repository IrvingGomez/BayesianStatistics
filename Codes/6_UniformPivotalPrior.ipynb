```python
from scipy.stats import uniform

import matplotlib.pyplot as plt
import numpy as np
```

```python
n = 10
a = -1
b = 1

sample = uniform.rvs(size=n, loc=a, scale=b-a)
```

```python
x_min = sample.min()
x_max = sample.max()
```

```python
def f(x,y):
    log_f = np.log(n*(n-1))+(n-1)*np.log(x_max-x_min)-(n+1)*np.log(y-x)
    return np.exp(log_f)
```

```python
x_vect = np.linspace(-1.1, x_min, 40)
y_vect = np.linspace(x_max, 1.1, 40)

X, Y = np.meshgrid(x_vect, y_vect)
Z = f(X,Y)
```

```python
plt.contour(X, Y, Z, 200, cmap='RdGy_r')
plt.colorbar()
```
